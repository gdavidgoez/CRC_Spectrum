# Midnight 00:06 / 2023-07-22

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
 
name_data1 = "output_163_2023-07-22 00_06_06.912"
file_data1 = '/home/david-imec/gnuradio_test/Tomorrowland_Air_T/dataset_Tomorrowland/'+ name_data1 +'.csv'

dataframe1 = pd.read_csv(file_data1, delimiter=',', header=None)
```

```python

frame = 800

sig1 = np.array(dataframe1.iloc[frame,0:].values,dtype=float)   # Signal data


# Plot the data
plt.figure(figsize=(15,6))
plt.plot(sig1)
plt.ylabel('PSD (dB)')
plt.xlabel('Freq (GHz)')
plt.title("Frequency domain")
plt.show()
```

```python
ft=7000
fre = 2400 + ft*(80e6/8192)/1e6

sig1 = np.array(dataframe1.iloc[:,ft].values,dtype=float)   # Signal data
# Plot the data
plt.figure(figsize=(15,6))
plt.plot(sig1)
plt.ylabel('PSD (dB)')
plt.xlabel('Time')
plt.title("Frequency evolution over time: "+str(fre))
plt.show()
```
